```python
class Roi:

    def __init__(self, monthly_income=0, monthly_expenses=0, monthly_cash_flow=0, annual_cash_flow=0, investments=0, roi=0):
        self.monthly_income = monthly_income
        self.monthly_expenses = monthly_expenses
        self.monthly_cash_flow = monthly_cash_flow
        self.annual_cash_flow = annual_cash_flow
        self.investments = investments
        self.roi = roi

        
    # Method for income
    def income(self):
        rental = float(input(" Monthly Rental Income = "))
        laundry = float(input(" Laundry Income = "))
        storage = float(input(" Storage Income = "))
        misc = float(input(" Miscelaneos = "))
        
        self.monthly_income = rental + laundry + storage + misc
        print(f" Monthly Income = ${self.monthly_income}")
        
    # Method for expenses
    def expenses(self):
        tax = float(input("Taxes = "))
        insurance = float(input("Insurance = "))
        electricity = float(input("Electricity = "))
        water = float(input("Water = "))
        sewer = float(input("Sewer = "))
        garbage = float(input("Garbage = "))
        gas = float(input("Gas = "))
        HOA = float(input("HOA = "))
        lawn_snow = float(input("Lawn and Snow = "))
        vacancy = float(input("Vacancy = "))
        repairs = float(input("Repairs = "))
        cap_ex = float(input("Capital Expenditures = "))
        property_mgmt = float(input("Property Management = "))
        mortgage = float(input("Mortgage = "))
        
        self.monthly_expenses = tax + insurance + electricity + water + sewer + garbage + gas + HOA + lawn_snow + vacancy + repairs + cap_ex + property_mgmt + mortgage
        print(f"Monthly Expenses = ${self.monthly_expenses}")
        
    # Method for cashflow
    def cash_flow(self):
        self.monthly_cash_flow = self.monthly_income - self.monthly_expenses
        print(f"Monthly Cash Flow = ${self.monthly_cash_flow}")
        
    # Method for cash on cash ROI
    def cash_on_cash_roi(self):
        self.annual_cash_flow = self.monthly_cash_flow * 12
        down_payment = float(input("Down Payment = "))
        closing_cost = float(input("Closing Costs = "))
        rehab_budget = float(input("Rehab Budget = "))
        misc_other = float(input("Miscelaneos Other = "))
        self.investments = down_payment + closing_cost + rehab_budget + misc_other
        self.roi = self.annual_cash_flow / self.investments
        self.roi_percent = self.roi * 100
        return self.roi_percent
    
    # Method 
    def roi_calculator(self):
        print("-"*50)
        print("Roi Calculator Result")
        print("-"*50)
        print(
            f""" 
    Total Monthly Income: ${self.monthly_income}
    Total Monthly Expenses: ${self.monthly_expenses}
    Monthly Cash Flow: ${self.monthly_cash_flow}
    Annual Cash Flow: ${self.annual_cash_flow}
    Total Investment: ${self.investments}

    Return On Investments: {self.roi_percent}%
    """
    ) 


    
def run_roi():
    print("-"*50)
    print("Roi Calculator")
    my_roi = Roi()
    print("-"*50)
    my_roi.income()
    my_roi.expenses()
    my_roi.cash_flow()
    my_roi.cash_on_cash_roi()
    my_roi.roi_calculator()


run_roi()
   
```
